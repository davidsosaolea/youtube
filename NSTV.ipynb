```python
# Eliminar las filas que contienen "MIRA EL PROGRAMA COMPLETO" en la columna "Descripci√≥n"
# df = df[~df['Descripci√≥n'].str.contains('MIRA EL PROGRAMA COMPLETO')]

# Mostrar las primeras filas del DataFrame despu√©s de la eliminaci√≥n
df.head()
df.Titulo
# Extraer el texto antes del " - " como "Titulo" y el texto despu√©s como "Programa"
df[['Titulo', 'Programa']] = df['Titulo'].str.extract(r'^(.) - (.)$')

# Mostrar las primeras filas del DataFrame despu√©s de la separaci√≥n
df.head()
df.Programa.unique()
# Crear una copia de la columna "Titulo" y rellenar los valores NaN en esa copia
titulo_filled = df['Titulo'].fillna('')

# Actualizar la columna "Titulo" con los valores rellenados
df['Titulo'] = titulo_filled


# Verificar si el t√≠tulo comienza con "reaccionando" y asignar "reaccionando" a la columna "Programa"
df.loc[df['Titulo'].str.contains('REACCIONANDO'), 'Programa'] = 'REACCIONANDO'
df.Programa.unique()
# Buscar filas con valores NaN en cualquier columna
filas_nan = df[df.isna().any(axis=1)]

# Mostrar las filas que contienen valores NaN
filas_nan
# Eliminar las filas que contienen valores NaN en cualquier columna
df_sin_nan = df.dropna()
df_sin_nan
```
