```python
# Eliminar las filas que contienen "MIRA EL PROGRAMA COMPLETO" en la columna "Descripci√≥n"
# Rellenar los valores NaN en la columna 'Programa' con una cadena vac√≠a

df['Descripci√≥n'] = df['Descripci√≥n'].fillna('')

df = df[~df['Descripci√≥n'].str.contains('MIRA EL PROGRAMA COMPLETO')]
```

```python
# Calcular la cantidad de valores nulos en cada columna
valores_nulos = df.isnull().sum()

print("Valores nulos por columna:")
print(valores_nulos)
```

```python
df
```

```python
# Crear un nuevo DataFrame con las columnas 'Titulo_pro' y 'Programa'
new_columns = df['Titulo'].str.split(r'\s*-\s*', n=1, expand=True)
new_columns.columns = ['Titulo_pro', 'Programa']

# Asignar el nuevo DataFrame al DataFrame original
df = pd.concat([df, new_columns], axis=1)
```

```python
df
```

```python
# Calcular la cantidad de valores nulos en cada columna
valores_nulos = df.isnull().sum()

print("Valores nulos por columna:")
print(valores_nulos)
```

```python
# Reemplazar los valores None en 'Programa' con los valores de 'Titulo'
df['Programa'] = df['Programa'].fillna(df['Titulo'])
```

```python
import unidecode

# Aplicar la funci√≥n unidecode para quitar las tildes de los valores en la columna 'Programa'
df['Programa'] = df['Programa'].apply(lambda x: unidecode.unidecode(x) if isinstance(x, str) else x)
```

```python
# Reemplazar los valores 'PDH' por 'PALABRA DE HINCHA' en la columna 'Programa'
df['Programa'] = df['Programa'].replace('PDH', 'PALABRA DE HINCHA')
df['Programa'] = df['Programa'].replace('SQC', 'SOLO QUEREMOS CONVERSAR')
```
